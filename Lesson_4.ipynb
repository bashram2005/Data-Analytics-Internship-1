```python
words = ['apple', 'banana', 'cherry', 'apple', 'date', 'banana', 'apple']

thisdict = {}

for fruit in words:
    if thisdict[fruit] == 1:
        thisdict[fruit] += 2
    else:
        thisdict[fruit] = 1
    
print(thisdict)
print(thisdict.values())
```

```python
words = ['apple', 'banana', 'cherry', 'apple', 'date', 'banana', 'apple']

thisdict = {}

for fruit in words:
    if fruit in thisdict:
       thisdict[fruit] += 1
    else:
        thisdict[fruit] = 1
    
print(thisdict)
```

```python
words = ['apple', 'banana', 'banana', 'apple', 'orange', 'orange', 'orange', 'apple', 'orange']

thisdict = {}

for fruit in words:
    if fruit in thisdict.keys():
       thisdict[fruit] += 1
    else:
        thisdict[fruit] = 1
    
print(thisdict)
```
