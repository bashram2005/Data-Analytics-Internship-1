```python
# ATM 4.7.26
class Bank_account:
    def __init__(self, name_of_owner, account_type, balance):
        self.name_of_owner = name_of_owner
        self.bank_name = 'bank of america'
        self.account_type = account_type
        self.balance = balance
        self.account_details = name_of_owner, account_type
        
    def printName_of_owner(self):
        print('Account owner name: ' + self.name_of_owner)
    def printBank_name(self):
        print('Name of bank: ' + self.bank_name)
    def printAccount_type(self):
        print('Account type:', self.account_type)
        
    def printAccount_details(self):
        print('Account details: ') 
        print('   Account owner name: ' + self.name_of_owner)
        print('   Name of bank: ' + self.bank_name)
        print('   Account type: ' + self.account_type)
        print('   Remaining balance:', self.balance)
        
    def withdraw(self, withdrawal_amount):
        if self.balance - withdrawal_amount >= 0:
            self.balance = self.balance - withdrawal_amount
            def atm_withdrawal(withdrawal_amount):

                outcome = 'Cash withdrawal:'
                dollar_bill = int(withdrawal_amount/100)
                outcome = (outcome + ' ' + str(dollar_bill)+ " of 100 dollar bills,")
                remainder = withdrawal_amount % 100

                dollar_bill = int(remainder/50)
                outcome = (outcome + ' ' + str(dollar_bill)+ " of 50 dollar bills,")
                remainder = remainder % 50

                dollar_bill = int(remainder/20) 
                outcome = (outcome + ' ' + str(dollar_bill)+ " of 20 dollar bills,")
                remainder = remainder % 20

                dollar_bill = int(remainder/10)
                outcome = (outcome + ' ' + str(dollar_bill)+ " of 10 dollar bills, and")
                remainder = remainder % 10

                dollar_bill = int(remainder/1)
                outcome = (outcome + ' ' + str(dollar_bill)+ " of 1 dollar bills.")

                return outcome

            atm_withdrawal_result = atm_withdrawal(withdrawal_amount)

            print(atm_withdrawal_result)

            self.printAccount_details()
        else:
            print('Insufficient funds')
            
    def deposit(self, amount_to_deposit):
        self.balance = self.balance + amount_to_deposit
        print('Cash deposit of:', amount_to_deposit)
        self.printAccount_details()

    def getBalance(self): 
        return self.balance
```

```python
charles_checking_account = Bank_account('Charles', 'Checking', 1000)

charles_savings_account = Bank_account('Charles', 'Savings', 1200)

charles_checking_account.printAccount_details()

charles_savings_account.printAccount_details()

charles_checking_account.withdraw(543)
charles_savings_account.deposit(200)
```
