```python
words = ['apple', 'banana', 'banana', 'apple', 'orange', 'orange', 'orange', 'apple', 'orange']

def counter(words):
    thisdict = {}

    for fruit in words:
        if fruit in thisdict.keys():
            thisdict[fruit] += 1
        else:
            thisdict[fruit] = 1
    return thisdict
    
dict_result = counter(words)

print(dict_result)
```

```python
# Dictionary function
num_list = [7,2,3,-1,-5,4,0,8]

def relative_max(num_list):

    relative_maximums = []

    first_element_index = 0

    if num_list[first_element_index] > num_list[first_element_index + 1]:
        relative_maximums.append(num_list[first_element_index])

    for i in range(1, len(num_list)-1):
        if num_list[i] > num_list[i-1] and num_list[i] > num_list[i+1]:
            relative_maximums.append(num_list[i])
        
    last_element_index = len(num_list) - 1

    if num_list[last_element_index] > num_list[last_element_index - 1]:
        relative_maximums.append(num_list[last_element_index])
    
    return relative_maximums

relative_max_result = 'relative maximums:',  relative_max(num_list)
    
print(relative_max_result)
```

```python
# Relative minimum and maximum function
num_list = [7,2,3,-1,-5,4,0,8]

def relative_min_and_max(num_list):

    relative_minimums = []
    
    relative_maximums = []

    first_element_index = 0

    if num_list[first_element_index] < num_list[first_element_index + 1]:
        relative_minimums.append(num_list[first_element_index]) 
        
    if num_list[first_element_index] > num_list[first_element_index + 1]:
        relative_maximums.append(num_list[first_element_index])

    for i in range(1, len(num_list)-1):
        if num_list[i] < num_list[i-1] and num_list[i] < num_list[i+1]:
            relative_minimums.append(num_list[i])
        elif num_list[i] > num_list[i-1] and num_list[i] > num_list[i+1]:
            relative_maximums.append(num_list[i])
        
    last_element_index = len(num_list) - 1

    if  num_list[last_element_index] < num_list[last_element_index - 1]:
        relative_minimums.append(num_list[last_element_index])

    if num_list[last_element_index] > num_list[last_element_index - 1]:
        relative_maximums.append(num_list[last_element_index])

    relative_minimums = 'relative minimums:', relative_minimums
    relative_maximums = 'relative maximums:', relative_maximums
    
    return relative_minimums, relative_maximums

    
relative_min_and_max_result = relative_min_and_max(num_list)

print(relative_min_and_max_result)
```

```python
# ATM function
withdrawal_amount=845

def atm_function(withdrawal_amount):

    outcome = 'here is:'
    dollar_bill = int(withdrawal_amount/100)
    outcome = (outcome + ' ' + str(dollar_bill)+ " of 100 dollar bills,")
    remainder = withdrawal_amount%100

    dollar_bill = int(remainder/50)
    outcome = (outcome + ' ' + str(dollar_bill)+ " of 50 dollar bills,")
    remainder = remainder%50

    dollar_bill = int(remainder/20) 
    outcome = (outcome + ' ' + str(dollar_bill)+ " of 20 dollar bills,")
    remainder = remainder%20

    dollar_bill = int(remainder/10)
    outcome = (outcome + ' ' + str(dollar_bill)+ " of 10 dollar bills, and")
    remainder = remainder%10

    dollar_bill = int(remainder/1)
    outcome = (outcome + ' ' + str(dollar_bill)+ " of 1 dollar bills.")

    return outcome

atm_function_result = atm_function(withdrawal_amount)

print(atm_function_result)
```

```python
# Palindrome function
user_input = 'madam'

def palindrome(user_input):

    index = 0
    is_palindrome = True

    while index < len(user_input) - index - 1:
        if user_input[index] == user_input[len(user_input) - index - 1]:
            index += 1
            user_input == is_palindrome
        else:
            is_palindrome = False
            break
            
    return is_palindrome

palindrome_result = palindrome(user_input)
        
print(palindrome_result)
```
