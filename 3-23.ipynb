```python
user_input = 'delirious'
index = 0
end_index = len(user_input) - 1
is_palindrome = True

while index < end_index:
    if user_input[index] == user_input[end_index]:
        index += 1
        end_index -= 1
        user_input == is_palindrome
    else:
        is_palindrome = False
        break
if is_palindrome:
    print(is_palindrome)
else:
    print('False')
```
