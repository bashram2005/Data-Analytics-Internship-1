```python
num_list = [7,2,3,-1,-5,4,0,8]

relative_maximum = []

first_element_index = 0

if num_list[first_element_index] > num_list[first_element_index + 1]:
    relative_maximum.append(num_list[first_element_index])

for i in range(1, len(num_list)-1):
    if num_list[i] > num_list[i-1] and num_list[i] > num_list[i+1]:
        relative_maximum.append(num_list[i])
        
last_element_index = len(num_list) - 1

if num_list[last_element_index] > num_list[last_element_index - 1]:
    relative_maximum.append(num_list[last_element_index])
    
print(relative_maximum)
```

```python
num_list = [1,-2,3,-1,-5,4,0,0]

relative_minimum = []

relative_minimum_index = []

for i in range(1, len(num_list) -2):
    if num_list[i] < num_list[i-1] and num_list[i] < num_list[i+1]:
        relative_minimum.append(num_list[i])
        relative_minimum_index.append(i)
print('relative minimum index:', relative_minimum_index)
print('relative minimums:     ', relative_minimum)
```
