```python
city = input("what city do you live in?")
temperature = float(input("what temperature is it?"))
is_raining = (input("is it raining? (Y/N)") == "Y")

print("In", city, "it is", temperature, "degrees and")
if is_raining:
    print("raining.")
else:
    print("not raining.")
```

```python
number = int(input("give me a number"))
if number == 0:
    print("number is zero")
elif number > 0:
    print("number is positive")
else: 
    print("number is negative")
```

```python
age = int(input("how old are you?"))
is_adult = (age >= 18)

if age <= 0:
    print("not born yet")
elif is_adult:
    print("adult")
else:
    print("minor")
```
